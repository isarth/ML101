```python
#I will try to generate a dataset with size 100 randomly with the above given probability
def dataset():
    import random
    data100=[]
    
    for i in range(100):
        prob= random.random()
        if prob <0.3:
            data100.append('a')
        elif prob >0.3 and prob <0.8:
            data100.append('b')
        else:
            data100.append('c')
    return data100        
```
